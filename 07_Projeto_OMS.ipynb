### Há um senso comum na sociedade que o uso do álcool em excesso pode vir a causar problemas psicológicos, além dos outros físicos, seria correto inferir que países com altas taxas de suicídio também apresentariam números elevados de consumo alcoólico? 


```python
# Limpeza de traços 


for i in df2["Suicide mortality ratek (per 100 000 population)"].index:
  if df2.loc[i,"Suicide mortality ratek (per 100 000 population)"] == "-":
             df2.loc[i,"Suicide mortality ratek (per 100 000 population)"] = np.NAN

for i in df2["Total alcohol per capita (≥ 15 years of age) consumptionl (litres of pure alcohol)"].index: 
  if df2.loc[i,"Total alcohol per capita (≥ 15 years of age) consumptionl (litres of pure alcohol)"] == "-":
             df2.loc[i,"Total alcohol per capita (≥ 15 years of age) consumptionl (litres of pure alcohol)"] = np.NAN
df2
```

## Qual a média de consumo de álcool entre as 10 nações com maior e menor índice de suicídio? 

```python
# Índice de Suicídio X Consumo de álcool


df2_maior = df2.iloc[:,0:4].sort_values(by='Suicide mortality ratek (per 100 000 population)', ascending=False)
df2_maior.drop(columns="Probability of dying from any of CVD, cancer, diabetes, CRD between age 30 and exact age 70k (%)",inplace=True)
df2_10ma= df2_maior.head(10)

df2_menor = df2.iloc[:, 0:4].sort_values(by='Suicide mortality ratek (per 100 000 population)', ascending=True)
df2_menor.drop(columns="Probability of dying from any of CVD, cancer, diabetes, CRD between age 30 and exact age 70k (%)",inplace=True)
df2_10me= df2_menor.head(10)
 
print(f"As 10 nações com MAIORES índices de suicídio, consomem em média {df2_10ma.iloc[:, 2].mean():.2f} litros de álcool puro per capta. ")
print(f"As 10 nações com MENORES índices de suicídio, consomem em média {df2_10me.iloc[:, 2].mean():.2f} litros de álcool puro per capta. ")
```

### Além do álcool, outras substâncias que geram dependência como o tabaco são vistas como pontos de atenção em caso de problemas psicológicos. Podemos visualizar a dependência variar conforme transita nos índices de suicídio?


```python
df3= pd.read_excel('dados/whs2022.xlsx', sheet_name=3, header=1)
df3=df3.drop([0,1,2])
df3.rename(columns={"Unnamed: 0": "Country"}, inplace= True)
df3.reset_index(drop= True, inplace=True)

df3
```

## Qual a diferença percentual e nominal de suicídios entre os países na parte superior e inferior do índice de uso do tabaco?

```python
#Limpeza e cruzamento dos dados

df3_tabaco = df3.iloc[:, 0:2]
df3_tabaco = df3_tabaco.replace('-','NaN')
df3_tabaco = df3_tabaco.drop(195) #Drop no totalizador da planilha
df3_tabaco.dropna(inplace=True)
df3_tabaco[df3_tabaco.columns[1]] = np.array(df3_tabaco[df3_tabaco.columns[1]], dtype = 'float16' ) 
df3_cruzado = pd.merge(df3_tabaco,df2_maior, how="left", right_on= "Country", left_on= "Country" )
df3_cruzado.drop(columns=df3_cruzado.columns[3], inplace=True)
df3_cruzado.dropna(inplace=True)
```

```python
# Os dez países que mais consomem

df3_mais_tabaco = df3_cruzado.sort_values(by=df3_cruzado.columns[1], ascending=False).head(10)
df3_mais_tabaco.reset_index(drop= True, inplace=True)

df3_mais_tabaco
```

```python
#Visualização dos dados de maior uso cruzado aos índices de suicídio referentes

ax_maior = sns.barplot(x=df3_mais_tabaco.columns[0], y = df3_mais_tabaco.columns[1], data = df3_mais_tabaco[0:10])
ax_maior2 = ax_maior.twinx()
sns.lineplot(x=df3_mais_tabaco.columns[0], y = df3_mais_tabaco.columns[2], data = df3_mais_tabaco[0:10])

ax_maior.set_xlabel('País')
ax_maior.set_ylabel('% Fumantes acima de 15 anos')
ax_maior.figure.set_size_inches(14, 10)
```

```python
# Os dez países que menos consomem

df3_menos_tabaco = df3_cruzado.sort_values(by=df3_tabaco.columns[1]).head(10)
df3_menos_tabaco.reset_index(drop= True, inplace=True)

df3_menos_tabaco
```

```python
#Visualização dos dados de menor uso cruzado aos índices de suicídio referentes

ax_menor = sns.barplot(x=df3_menos_tabaco.columns[0], y = df3_menos_tabaco.columns[1], data = df3_menos_tabaco)
ax_menor2 = ax_menor.twinx()
sns.lineplot(x=df3_menos_tabaco.columns[0], y = df3_menos_tabaco.columns[2], data = df3_menos_tabaco[0:10])

ax_menor.set_xlabel('País')
ax_menor.set_ylabel('% Fumantes acima de 15 anos')
ax_menor.figure.set_size_inches(14, 10)
```

```python
#Respondendo o questionamento anterior com as bases construídas

dif_percentual = (df3_menos_tabaco[df3_menos_tabaco.columns[2]].mean()/df3_mais_tabaco[df3_mais_tabaco.columns[2]].mean())*100
dif_nominal = df3_mais_tabaco[df3_mais_tabaco.columns[2]].sum()-df3_menos_tabaco[df3_menos_tabaco.columns[2]].sum()

print(f"Os países com maior uso de tabaco tem um índice de suicídios {dif_percentual:.2f}% superior aos países com menor uso.")
print(f"A diferença percentual representa mais {dif_nominal:.2f} casos entre o ranking total.")
```

### Violência contra a mulher 

```python
import pandas as pd
suicidio = pd.read_csv('dados/suicide_rates_10.csv', usecols=['Indicator','ParentLocation' , 'Location', 'Period', 'Dim1', 'Dim2', 'Value' ])
violencia= pd.read_excel('dados/whs2022.xlsx', sheet_name=4, header = 0,)
```

```python
suicidio = suicidio.sort_values(['Value'], ascending=False)
suicidio = suicidio.rename(columns={"Dim1":"Sex","Dim2":"Age"})
suicidio.columns.name = 'Id'
suicidio = suicidio[suicidio['Sex']=='Female']
suicidio = suicidio.query("Age == '15-24 years'| Age == '25-34 years of age' | Age == '35-44 years of age' | Age == '45-54 years'")
suicidio = suicidio.drop(columns=['Indicator'])
suicidio = suicidio.rename(columns={"Value":"% of suicides per 100,000 inhabitants"})
suicidio.head(10)
```

```python
violencia = violencia[['Unnamed: 0','Unnamed: 9']]
violencia = violencia.drop([0,1,2,3])
violencia.columns.name = 'Id'
violencia = violencia.rename(columns={"Unnamed: 0":"Location","Unnamed: 9":"Value"})
violencia.index = range(1,violencia.shape[0]+1)
violencia = violencia.drop([195,196,197,198,199,200,201,202,203,204])
filtro  = violencia['Value'] != '-'
violencia = violencia[filtro]
violencia = violencia.sort_values(['Value'], ascending=False)
violencia.index = range(1,violencia.shape[0]+1)
violencia = violencia.rename(columns={"Value":"% of lifetime domestic violence"})
violencia.head(10)
```

```python
dadoscruzados = pd.merge(violencia,suicidio, on = "Location")
dadoscruzados.head(24)
```

### É possível perceber que, com exceção de Bangladesh os outros paises que estão no ranking de violencia doméstica e que também estão no ranking de suicidio são predominantemente da Oceania.
### E também que em Kiribati e Fiji, que lideram em número de violencia doméstica, as mulheres mais jovens estão se suicidando mais com o passar do tempo.

```python
dadoscruzados.tail(24)
```

### Já nos países com menor indíce de violencia doméstica, com exceção de Singapura, pode-se perceber que mulheres mais velhas cometem mais suicidios.
